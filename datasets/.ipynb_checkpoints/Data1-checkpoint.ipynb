```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv("./bodyfind.csv")
df.sample(5)
```

```python
cols = list(df.columns)
```

```python
%matplotlib inline
```

```python
for lab in cols[:-1]:
    plt.hist(df[df["Dosha"]==0][lab],color='b',alpha = 0.6, density = True, label='vata')
    plt.hist(df[df["Dosha"]==1][lab],color='r',alpha = 0.6, density = True, label='pitta')    
    plt.hist(df[df["Dosha"]==2][lab],color='g',alpha = 0.6, density = True, label='kapha')  
    plt.xlabel(lab)
    plt.ylabel('Probability')
    plt.title(lab)
    plt.legend()
    plt.show()
    
```

```python
X = df.drop('Dosha', axis ='columns')
y = df['Dosha']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2);
X_train.shape,X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, cmap='Blues')
plt.xlabel("Truth")
plt.ylabel('Predicted')
plt.title("Confusion Matrix")
plt.tight_layout()
plt.show()
```

```python
from joblib import load, dump
dump(model,'./trained_model')
```
